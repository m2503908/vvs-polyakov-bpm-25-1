```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())
```

```python
titanic_data
```

```python
titanic_data = titanic_data.drop(["Name", "Ticket", "Cabin"], axis=1) # убираем столбцы Name, Cabin и Ticket, так как они неинформативны для анализа данных
```

```python
titanic_data['Sex'] = titanic_data['Sex'].replace({"male": 0, "female": 1})
```

```python
titanic_data.groupby(titanic_data["Embarked"]).agg({"PassengerId": "size"})
```

```python
titanic_data['Embarked'] = titanic_data['Embarked'].replace({"Q": 46, "C": 102, "S": 270}) 
# поскольку мы заменяем категориальный на численный, то сделаем это разумно, меняя не просто на 0 1 2, а на количество
# каждого такого элемента в таблице, для сохранения какой-либо логики
```
